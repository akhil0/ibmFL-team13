```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import seaborn as sns
from matplotlib.ticker import ScalarFormatter, FormatStrFormatter
```

```python
booking_data = pd.read_csv('data.csv')
```

```python
booking_data.head()
```

```python
booking_data['departure_time'] =  pd.to_datetime(booking_data['departure_time'], format='%Y-%m-%d %H:%M:%S')
booking_data['hour'] = booking_data.departure_time.apply(lambda x: x.hour)
booking_data['minute'] = booking_data.departure_time.apply(lambda x: x.minute)
booking_data['month'] = booking_data.departure_time.apply(lambda x: x.month)
booking_data['sday'] = booking_data.departure_time.apply(lambda x: x.day)
```

```python
sns.set_style('whitegrid')
cp = sns.color_palette("BrBG")
ax = sns.countplot(x="hour", data=booking_data, color="dodgerblue")
ax.yaxis.set_major_formatter(FormatStrFormatter('%.0f'))
ax.set_xlabel('Hour of the Day', fontsize = 12)
ax.set_ylabel('Count of flight times', fontsize = 12)
ax.set_title('Flying times by hour', fontsize=16)
ax.tick_params(labelsize = 8)
plt.show()
```

```python
sns.set_style('whitegrid')
ax = sns.countplot(x="day", data=booking_data, color="dodgerblue")
ax.yaxis.set_major_formatter(FormatStrFormatter('%.0f'))
ax.set_xlabel('Day of the week', fontsize = 12)
ax.set_ylabel('Count of flight Times', fontsize = 12)
ax.set_title('Flying times by Day', fontsize=16)
ax.tick_params(labelsize = 8)
plt.show()
```

```python
monday_data = booking_data[booking_data['day']=='monday']
wednesday_data = booking_data[booking_data['day']=='wednesday']
friday_data = booking_data[booking_data['day']=='friday']
```

```python
def show_plt(metadata,day):
    sns.set_style('whitegrid')
    cp = sns.color_palette("BrBG")
    ax = sns.countplot(x="to", data=metadata, color="dodgerblue")
    ax.yaxis.set_major_formatter(FormatStrFormatter('%.0f'))
    ax.set_xlabel('Diffferent Locations Brianne flying to', fontsize = 12)
    ax.set_ylabel('Count of flights', fontsize = 12)
    ax.set_title('Flight Counts by Location on '+day, fontsize=16)
    ax.tick_params(labelsize = 8)
    plt.show()
```

```python
show_plt(monday_data,'Monday')
```

```python
show_plt(wednesday_data,'Wednesday')
```

```python
show_plt(friday_data,'Friday')
```
